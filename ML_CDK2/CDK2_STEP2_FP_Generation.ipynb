```python
import glob
xml_files = glob.glob("*.xml")
xml_files.sort()
xml_files
```

```python
FP_list = ['AtomPairs2DCount',
 'AtomPairs2D',
 'EState',
 'CDKextended',
 'CDK',
 'CDKgraphonly',
 'KlekotaRothCount',
 'KlekotaRoth',
 'MACCS',
 'PubChem',
 'SubstructureCount',
 'Substructure']
```

```python
fp = dict(zip(FP_list, xml_files))
fp
```

```python
from padelpy import padeldescriptor

fingerprint = 'Substructure'

fingerprint_output_file = ''.join([fingerprint,'.csv']) #Substructure.csv
fingerprint_descriptortypes = fp[fingerprint]

padeldescriptor(mol_dir='molecule.smi',
                d_file=fingerprint_output_file, #'Substructure.csv'

                descriptortypes= fingerprint_descriptortypes,
                detectaromaticity=True,
                standardizenitro=True,
                standardizetautomers=True,
                threads=2,
                removesalt=True,
                log=True,
                fingerprints=True)
```

```python
df6sub = pd.read_csv(fingerprint_output_file)
df6sub.head(3)
```

```python
len(df6sub)
```

```python
dfsub = pd.concat([df10.reset_index(),df6sub .reset_index()], axis=1)
dfsub.head(3)
```

```python
#dfsub['targetName'].unique()
```

```python
len(dfsub)
```

```python
#dfsub['targetSource'].unique()
```

```python
dfsub.columns
```

```python
dfsub1= dfsub.drop(['index', 'molecule_chembl_id', 'canonical_smiles','units', 'standard_value', 'index', 'Name'], axis=1)
```

```python
dfsub1.to_csv('substructure_CDK2.csv')
```
